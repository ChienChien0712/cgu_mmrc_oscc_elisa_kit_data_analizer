```python
import pandas as pd
import numpy as np
```

```python
f1 = pd.read_excel('data/demo_4gp.xlsx')
f1.head()
```

```python
gp = []
gp_name = []
for i in list(range(len(f1))):
    if f1.at[i,'group 4'] == 'A':
        gp.append(1)
        gp_name.append('HC')
    elif f1.at[i,'group 4'] == 'B':
        gp.append(2)
        gp_name.append('OPMD I')
    elif f1.at[i,'group 4'] == 'C':
        gp.append(3)
        gp_name.append('OPMD II')        
    elif f1.at[i,'group 4'] == 'D':
        gp.append(4)
        gp_name.append('OSCC')
    else:
        gp.append(99)
        gp_name.append('NA')     
        
f1['分群'] = gp  
f1['分群名'] = gp_name
```

```python
f1.head()
```

```python
Neg_Name = [1,2,3]
pos_name = 4
gp_name = ['HC','OPMD I','OPMD II','OSCC']
conc_name = 'ELISA'

#writer
for neg_name in Neg_Name:
    writer = pd.ExcelWriter('./results/ROC_%svs%s_%s.xlsx'%(gp_name[neg_name-1],gp_name[pos_name-1],conc_name))
    neg_conc = f1[(f1['分群']==neg_name)][conc_name]
    pos_conc = f1[(f1['分群']==pos_name)][conc_name]
    neg_count = len(neg_conc)
    pos_count = len(pos_conc)

    all_conc = sorted(list(np.unique(list(neg_conc)+list(pos_conc))))
    cut_max = max([max(neg_conc), max(pos_conc)])
    cut_start = -1

    CutOff = []
    Sen = []
    Spe = []

    sen_start = -1
    spe_start = -1

    for i in range(len(all_conc)-1):
        sen = sum(pos_conc >= cut_start)/pos_count
        spe = sum(neg_conc < cut_start)/neg_count

        CutOff.append(cut_start)
        Sen.append(sen)
        Spe.append(spe)

        cut_start = (all_conc[i]+all_conc[i+1]) / 2
        sen_start = sen
        spe_start = spe

    CutOff.append(cut_start )
    Sen.append(sum(pos_conc >= cut_start )/pos_count)
    Spe.append(sum(neg_conc < cut_start )/neg_count)    
    CutOff.append(cut_max+1)
    Sen.append(sum(pos_conc >= cut_max+1)/pos_count)
    Spe.append(sum(neg_conc < cut_max+1)/neg_count)

    SS = np.array(Sen) + np.array(Spe)
    one_minus_spe = 1- np.array(Spe)

    ColName = ['Positive if Greater Than or Equal To','Sensitivity','Specificity','1 - Specificity','S+S']

    ROC = pd.DataFrame(np.matrix([CutOff,Sen,Spe,one_minus_spe,SS]).T)
    ROC.columns = ColName
    #best cut-off
    #1. max S+S
    SS = list(SS)
    cutoff_SSmax_idx = SS.index(max(SS))

    #2. round(spe,1)>=0.9 & max(sen)
    submax_sen = max(ROC[ROC['Specificity']>=0.89]['Sensitivity'])
    cutoff_spe09_idx = ROC[(ROC['Specificity']>=0.89) & (ROC['Sensitivity']==submax_sen)].index[0]

    CutOff_max = [CutOff[cutoff_SSmax_idx],CutOff[cutoff_spe09_idx]]
    Sen_max = [Sen[cutoff_SSmax_idx],Sen[cutoff_spe09_idx]]
    Spe_max = [Spe[cutoff_SSmax_idx],Spe[cutoff_spe09_idx]]
    one_minus_spe_max = [one_minus_spe[cutoff_SSmax_idx],one_minus_spe[cutoff_spe09_idx]]
    SS_max = [SS[cutoff_SSmax_idx],SS[cutoff_spe09_idx]]
    '''
    #add cutoff = 200
    CutOff_max.append(200)
    Sen_max.append(sum(pos_conc >= 200)/pos_count)
    Spe_max.append(sum(neg_conc < 200)/neg_count)    
    one_minus_spe_max.append(1-sum(neg_conc < 200)/neg_count)
    SS_max.append(sum(pos_conc >= 200)/pos_count + sum(neg_conc < 200)/neg_count)
    #add cutoff = 100
    CutOff_max.append(100)
    Sen_max.append(sum(pos_conc >= 100)/pos_count)
    Spe_max.append(sum(neg_conc < 100)/neg_count)    
    one_minus_spe_max.append(1-sum(neg_conc < 100)/neg_count)
    SS_max.append(sum(pos_conc >= 100)/pos_count + sum(neg_conc < 100)/neg_count)    
    '''
    #95%CI
    Lower_Sen = []
    Upper_Sen = []
    Lower_Spe = []
    Upper_Spe = []
    for c in CutOff_max:
        A = sum(pos_conc >= c)
        B = sum(neg_conc >= c)
        C = sum(pos_conc < c)
        D = sum(neg_conc < c)
        Q1_sen = 2*A + 1.96**2
        Q2_sen = 1.96*np.sqrt(1.96**2+(4*A*C/(A+C)))
        Q3_sen = 2*(A+C+1.96**2)
        lower_sen = (Q1_sen-Q2_sen)/Q3_sen
        upper_sen = (Q1_sen+Q2_sen)/Q3_sen
        Q1_spe = 2*D + 1.96**2
        Q2_spe = 1.96*np.sqrt(1.96**2+(4*B*D/(B+D)))
        Q3_spe = 2*(B+D+1.96**2)
        lower_spe = (Q1_spe-Q2_spe)/Q3_spe
        upper_spe = (Q1_spe+Q2_spe)/Q3_spe       
        
        Lower_Sen.append(lower_sen)
        Upper_Sen.append(upper_sen)
        Lower_Spe.append(lower_spe)
        Upper_Spe.append(upper_spe)
    
    ROC_max = pd.DataFrame(np.matrix([CutOff_max,Sen_max,Spe_max,one_minus_spe_max,SS_max,Lower_Sen,Upper_Sen,Lower_Spe,Upper_Spe]).T)
    ROC_max.columns = ColName + ['95% CI of Sensitivity (lower)','95% CI of Sensitivity (upper)','95% CI of Specificity (lower)','95% CI of Specificity (upper)']
    ROC_max.index = ['max S+S','round(spe>=0.9)&max(sen)']

    
    
    
    
    #write to excel
    ROC.to_excel(writer,'ROC',index=False)
    ROC_max.to_excel(writer,'Best Cut-off')
    writer.save()
```
